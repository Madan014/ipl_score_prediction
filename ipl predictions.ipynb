# feature engineering 

```python
# ytrain test splits
x=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]
```

```python
x.head(1)
```

```python
y.head(1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
x_train.head(2)
```

# one hot encoding on the batting_teram,bowling_team and city

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([
    ('trf1',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
],remainder='passthrough')
print(" transform columns sucessfully")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
print(" imported sucessfully")
```

# logistic regressions algorithm

```python
pipe_logistic=Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
print(" make the pipeline sucessfully")
```

# naive bayes algorithms

```python
from sklearn.naive_bayes import GaussianNB
pipe_naive_bayes=Pipeline(steps=[
    ('step1',trf),
    ('step2',GaussianNB())
])
print(" make the pipeline sucessfully for the guassina naive bayes ")
```

# random forest algorithm

```python
from sklearn.ensemble import RandomForestClassifier
pipe_random_forest=Pipeline(steps=[
    ('step1',trf),
    ('step2',RandomForestClassifier())
])
print(" make the pipeline sucessfully for the random forests ")
```

```python
# fit both the pipelines 
pipe_random_forest.fit(x_train,y_train)
pipe_logistic.fit(x_train,y_train)
pipe_naive_bayes.fit(x_train,y_train)
print("fit both the pipelines")
```

# check the accuracy of the both algorithms

```python
y_pred_logistic=pipe_logistic.predict(x_test)
y_pred_naive=pipe_naive_bayes.predict(x_test)
y_pred_random_forest=pipe_random_forest.predict(x_test)
print(" sucess")
```

```python
from sklearn.metrics import accuracy_score
print("Naive Bayes : ",accuracy_score(y_test,y_pred_naive))
print("logistic Regression : ",accuracy_score(y_test,y_pred_logistic))
print("Random Forest Algorithm : ",accuracy_score(y_test,y_pred_random_forest))
```
